```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
```

```python
allrecp=pd.read_csv("recipes_links.csv",index_col=0)
```

```python
allrecp.columns=["recp_link"]
```

```python
recplst=list(allrecp['recp_link'])
```

```python
recplst[0]
```

```python
fdf=pd.DataFrame()
```

```python
for i in range(len(recplst)):
    url=recplst[i]
    response = requests.get(url)
    soup = bs(response.content, 'html.parser') 
    # rev_div = soup.findAll("div",attrs={"class","block-post-listing__inner"}) 
    # links=re. findall("href=[\"https\'](.*?)[\"\']", str(rev_div[0]))

    recipe_name=soup.h1.string # recipe name

    auth=soup.find("p",attrs={"class","entry-author"})
    try:
        auth_name=auth.findAll("a")[1].text #Author Name
    except:
        auth_name=''
    try:
        recipe_description=soup.find("div",attrs={"class","wprm-recipe-summary wprm-block-text-normal"}).text # recipe_description
    except:
        recipe_description=''
    try:
        recipe_rating=soup.find("div",attrs={"class","wprm-recipe-rating wprm-user-rating wprm-recipe-rating-inline wprm-user-rating-allowed"})['data-average'] # recipe_rating
    except:
        recipe_rating=''
    try:
        nos_rating=soup.find("div",attrs={"class","wprm-recipe-rating wprm-user-rating wprm-recipe-rating-inline wprm-user-rating-allowed"})['data-count'] # nos_rating
    except:
        nos_rating=''
    try:
        prep_time=soup.find("span",attrs={"class","wprm-recipe-details wprm-recipe-details-minutes wprm-recipe-prep_time wprm-recipe-prep_time-minutes"}).text # recipe_rating
        prep_timeunit=soup.find("span",attrs={"class","wprm-recipe-details-unit wprm-recipe-details-minutes wprm-recipe-prep_time-unit wprm-recipe-prep_timeunit-minutes"}).text # recipe_rating
        prep_time=prep_time+" "+prep_timeunit
    except:
        prep_time=''
    try:
        cook_time=soup.find("span",attrs={"class","wprm-recipe-details wprm-recipe-details-minutes wprm-recipe-cook_time wprm-recipe-cook_time-minutes"}).text # CookTime
        cook_time_unit=soup.find("span",attrs={"class","wprm-recipe-details-unit wprm-recipe-details-minutes wprm-recipe-cook_time-unit wprm-recipe-cook_timeunit-minutes"}).text # recipe_rating
        cook_time=cook_time+" "+cook_time_unit
    except:
        cook_time=''

    #TotalTime
    try:
        total_time=soup.find("span",attrs={"class","wprm-recipe-details wprm-recipe-details-minutes wprm-recipe-total_time wprm-recipe-total_time-minutes"}).text 
        total_time_unit=soup.find("span",attrs={"class","wprm-recipe-details-unit wprm-recipe-details-minutes wprm-recipe-total_time-unit wprm-recipe-total_timeunit-minutes"}).text
        total_time=total_time+" "+total_time_unit
    except:
        total_time=''
    #Cuisine 
    try:
        cuisine=soup.find("span",attrs={"class","wprm-recipe-cuisine wprm-block-text-bold"}).text
    except:
        cuisine=''

    #Course
    try:
        course=soup.find("span",attrs={"class","wprm-recipe-course wprm-block-text-bold"}).text
    except:
        course=''

    #Diet
    try:
        suitablefordiet=soup.find("span",attrs={"class","wprm-recipe-suitablefordiet wprm-block-text-bold"}).text
    except:
        suitablefordiet=''

    #DifficultyLevel
    try:
        difficulty=soup.find("span",attrs={"class","wprm-recipe-difficulty wprm-block-text-bold"}).text
    except:
        difficulty=''

    #servings
    try:
        servings=soup.find("span",attrs={"class","wprm-recipe-servings-with-unit"}).text #amount
    except:
        servings=''
    ind=soup.find("div",attrs={"class","wprm-recipe-ingredient-group"})
    #ingredients
    ingredients=[]
    try:
        for item in ind:
            try:
                ingds=item.h4.text.replace('▢',',')
            except:
                ingds='Main ingredients'
            ingds1=item.text.split('▢')[1:]
            ingredients.append({ingds:ingds1})
    except:
        ingredients=[]
    #instructions
    try:
        #instructions
        ins=soup.findAll("div",attrs={"class","wprm-recipe-instruction-group"})
        inslst=[]
        for item in ins:
            try:
                ins1=item.h4.text
            except:
                ins1='Preparation'
            ins2=item.text
            inslst.append({ins1:[ins2]})
    except:
        inslst=[]

    data=pd.DataFrame([recipe_name,auth_name,
    recipe_description,recipe_rating,
    nos_rating,prep_time,cook_time,
    total_time,cuisine,course,suitablefordiet,difficulty,
    servings,ingredients,inslst]).T

    data.columns=['recipe_name','auth_name','recipe_description','recipe_rating','nos_rating','prep_time','cook_time',
                  'total_time','cuisine','course','suitablefordiet','difficulty','servings','ingredients','inslst']

    fdf=pd.concat([fdf,data]).reset_index(drop=True)
    print("Recipe No : "+str(i))
```

```python
fdf.to_csv('recipes1.csv')
```
