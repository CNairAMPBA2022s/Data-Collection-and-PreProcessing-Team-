```python
import pandas as pd
import requests
from bs4 import BeautifulSoup as bs
```

```python
allrecp=pd.read_csv("C:/Users/Cherry/Desktop/Data Science/1. ISB_AMPBA/Term 1/Data Cleaning and Pre-Processing/Group Project/Manoj test 1.csv",index_col=0)
```

```python
allrecp
```

```python
allrecp.columns=["recp_link"]
```

```python
recplst=list(allrecp['recp_link'])
```

```python
recplst[0]
```

```python
fdf=pd.DataFrame()
```

```python
for i in range(len(recplst)):
    url=recplst[i]
    response = requests.get(url)
    soup = bs(response.content, 'html.parser') 
    # rev_div = soup.findAll("div",attrs={"class","block-post-listing__inner"}) 
    # links=re. findall("href=[\"https\'](.*?)[\"\']", str(rev_div[0]))

    recipe_name=soup.find('h1', class_='heading__title').text # recipe name

    auth=soup.find("p",attrs={"class","entry-author"})
    try:
        auth_name=soup.find('a', class_ = 'mntl-attribution__item-name').text #Author Name
    except:
        auth_name=''
    try:
        recipe_description=soup.find('h2', class_ = 'heading__subtitle').text # recipe_description
    except:
        recipe_description=''
    try:
        prep_time=soup.find_all('span', class_ = 'meta-text__data')[0].text
    except:
        prep_time=''
    try:
        cook_time = soup.find_all('span', class_ = 'meta-text__data')[1].text
    except:
        cook_time=''

    #TotalTime
    try:
        total_time = soup.find_all('span', class_ = 'meta-text__data')[2].text
    except:
        total_time=''
    #servings
    try:
        servings = soup.find_all('span', class_ = 'meta-text__data')[3].text
    except:
        servings=''
    
    ingre = soup.find_all('li', class_ = 'simple-list__item js-checkbox-trigger ingredient text-passage')
    #ingredients
    ingredients=[]
    try:
        for li_tag in ingre:
            b = li_tag.text
            ingredients.append(b)
    except:
        ingredients=[]
        
     #ingredients2
    ingre2 = soup.find_all('li', class_ = 'structured-ingredients__list-item')
    ingredients2=[]
    try:
        for li_tag in ingre2:
            ingredient2 = []
            b2 = li_tag.text
            ingredient2.append(b2)
    except:
        ingredients2=[]
    #instructions
    try:
        #instructions
        method = soup.find_all('p', class_ = 'comp mntl-sc-block mntl-sc-block-html')
        inslst=[]
        for p_tag in method:
            a = p_tag.text
            inslst.append(a)
    except:
        inslst=[]

    data=pd.DataFrame([recipe_name,auth_name,
    recipe_description,prep_time,cook_time,
    total_time,servings,ingredients,ingredients2,inslst]).T

    data.columns=['recipe_name','auth_name','recipe_description','prep_time','cook_time',
                  'total_time','servings','ingredients','ingredients2','inslst']

    fdf=pd.concat([fdf,data]).reset_index(drop=True)
    print("Recipe No : "+str(i))
```
