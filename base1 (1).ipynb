```python
import pandas as pd
import requests
import re
from bs4 import BeautifulSoup as bs
recp=152
allrecp=[]
for i in range(152):
    base_url = "https://www.vegrecipesofindia.com/recipes/?fwp_paged="
    url=base_url+str(i)
    response = requests.get(url)
    soup = bs(response.content, 'html.parser') 
    rev_div = soup.findAll("div",attrs={"class","block-post-listing__inner"}) 
    links=re. findall("href=[\"https\'](.*?)[\"\']", str(rev_div[0]))
    print(links)
    links=[x for x in links if not x.startswith('#')]
    print(links)
    links=list(set(links))
    allrecp.extend(links)
    print(len(allrecp))
```

```python
import pandas as pd
pd.DataFrame(allrecp).to_csv("recipes_links.csv")
```
